```python

def status_def(i):
    if 'Mr.' or 'Master.' in i:
        return 1
    if 'Mrs.' or 'Dona.' in i:
        return 3
    if 'Miss.' in i:
        return 2
    else:
        return 0
train['status']=[status_def(i) for i in train['Name'].values]
```

```python
def deck_def(i):
    if i == i:
        #print(i[0])
        deck_num='ABCDEFGT'
        return deck_num.index(str(i[0]))
    else:
        return 9
def numder_def(i):
    flag=0
    if i == i and i != '':
        ret=''
        for j in i:
            if j in'0123456789':
                ret+=j
                flag=1
            elif j!=i[0] and flag==1:
                return ret
        if ret=='':
            return 0
        else:
            return ret
    else:
        return 0
    
train['deck']=[deck_def(i) for i in train['Cabin'].values]
train['number']=[numder_def(i) for i in train['Cabin'].values]

def adult(i):
    if i<18:
        return 0
    else:
        return 1
train['adult']=[adult(i) for i in train['Age'].values]

train
```
