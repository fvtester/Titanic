```python

def status_def(i):
    if 'Mr.' or 'Master.' in i:
        return 1
    if 'Mrs.' or 'Dona.' in i:
        return 3
    if 'Miss.' in i:
        return 2
    else:
        return 0
train['status']=[status_def(i) for i in train['Name'].values]
```

```python
def deck_def(i):
    if i == i:
        #print(i[0])
        deck_num='ABCDEFGT'
        return deck_num.index(str(i[0]))
    else:
        return 9

train['deck']=[deck_def(i) for i in train['Cabin'].values]


def adult(i):
    if i<18:
        return 0
    else:
        return 1
train['adult']=[adult(i) for i in train['Age'].values]

train
```

```python
y_train=train['Survived']
```

```python
X_train=train.drop(['Survived', 'Name', 'Ticket', 'Cabin'], axis=1)
Ag_mean=X_train['Age'].mean()
print(Ag_mean)
X_train['Age']=X_train['Age'].fillna(Ag_mean)
X_train['Embarked']=X_train['Embarked'].fillna('N')
```
