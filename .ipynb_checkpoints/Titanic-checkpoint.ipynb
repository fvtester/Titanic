```python
import pandas as pd
```

```python
train=pd.read_csv('C:\\Users\\pi.karasev\\Documents\\titanic\\train.csv')
```

```python
train.head(20)
```

```python
y_train=train['Survived']
```

```python
X_train=train.drop(['Survived', 'Name', 'Ticket', 'Cabin'], axis=1)
X_train=X_train.fillna('0')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one=OneHotEncoder()
columns=['Sex']
X_train_encoded=one.fit_transform(X_train[columns])
pd.DataFrame(X_train_encoded)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(X_train['Sex'])
X_train['Sex'] = le.transform(X_train['Sex'])
X_train.head(20)
```

```python
le1 = LabelEncoder()
le1.fit(X_train['Embarked'])
X_train['Embarked'] = le1.transform(X_train['Embarked'])
X_train
```

```python
forest=RandomForestClassifier()
```

```python
forest.fit(X_train, y_train)
```
